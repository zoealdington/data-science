```python
def mentions_company(parsed):
    parsed = nlp_toolkit(parsed)
    for entity in parsed.ents:
        return ent.label_=='ORG'
```

```python
mentions_company(u'This sentence mentions no companies')
```

```python
mentions_company(u'This sentence mentions Google')
```

```python
def mentions_specific_company(parsed, company='Google'):
    parsed = nlp_toolkit(parsed)
    for ent in parsed.ents:
        return (ent.label_=='ORG' and ent.text==company)
```

```python
mentions_specific_company(parsed=u'This sentence mentions Apple',company='Google')
```

```python
mentions_specific_company(parsed=u'This sentence mentions Google')
```
